[4] 독립변수와 종속변수 및 파라미터 초기화

```python
# x : 독립변수([1, x]), y : 종속변수
x = np.array([[1,1], [1,2], [1,3]])
y = np.array([2, 3, 4])

# 선형회귀의 계수
w = np.array([0.3, 0.5])
```

[5] 선형회귀 객체 생성 및 학습

```python
linReg = LinRegression(1, w)
linReg.fit(x, y, nSamples, nEpochs, 0.01)
```
