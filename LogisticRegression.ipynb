[2] 로지스틱 함수 및 이진 교차 엔트로피 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def binary_CE(y, y_hat):
    return -y * np.log(y_hat) - (1-y) * np.log(1-y_hat)
```

[4] 로지스틱 회귀를 위한 설정

[5] 독립변수와 종속변수 및 파라미터 초기화

```python
# x : 독립변수([1, x]), y : 종속변수
x = np.array([[1,1, 1], [1, 2, 1], [1, 1, 2], [1, 2, 3], [1, 3, 2], [1, 3, 3]])
y = np.array([0, 0, 0, 1, 1, 1])

# 로지스틱 회귀의 계수
w = np.array([0.5, -0.5, 1.0])
```

[6] 로지스틱 회귀 객체 생성 및 학습

```python
logisticReg = LogisticRegression(2, w)
logisticReg.fit(x, y, nSamples, nEpochs, 0.01)
```
