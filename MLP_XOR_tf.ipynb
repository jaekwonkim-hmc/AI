[1] 필요한 패키지 불러오기

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras.optimizers import SGD
import numpy as np
```

[2] XOR를  구현하기 위한 모델 구성

```python
xor = keras.Sequential()
xor.add(keras.Input(shape=(2,)))
xor.add(Dense(4, activation='tanh'))
xor.add(Dense(1, activation='sigmoid'))
xor.summary()
```

[3] 독립변수와 종속변수

```python
# X : 독립변수, y : 종속변수
X = np.array([[0, 0],
              [1, 0],
              [0, 1],
              [1, 1]], dtype=np.float32)
y = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

[4] 모델 컴파일 및 훈련

```python
xor.compile(optimizer=SGD(learning_rate=0.1, momentum=0.9), loss='mse', metrics=['accuracy'])
steps = 20
format_str = 'epoch : {:4d}  손실 : {:.6f}, 정확도 : {:.6f}'
for i in range(50):
    xor.fit(X, y, verbose=0, epochs=steps)
    tr_loss, tr_acc = xor.evaluate(X, y, verbose=0)
    print(format_str.format((i+1)*steps, tr_loss, tr_acc))
```

[5] 훈련된 모델을 이용한 추론

```python
print(xor.predict(X))
```
